
## Mean Nightly Accommodation Costs and Daily Day-to-Day Expenditure of Foreign Resident Overnight Visitors
### Dataset taken from: https://data.cso.ie/


```python
import pandas as pd

# Load the CSV file into a DataFrame again

data_df = pd.read_csv("D/N_Foreign_Resident_Overnight_Visitors.csv")

# Display the first few rows of the data to understand its structure
data_df.head()
```

```python
import matplotlib.pyplot as plt

# Filter data for Mean Nightly Accommodation Costs for the month of "2023 April"
filtered_data = data_df[(data_df["Statistic Label"].str.contains("Mean Nightly Accommodation Costs")) & 
                        (data_df["Monthly"] == "2023 April")]

# Plotting the data
plt.figure(figsize=(12, 6))
plt.bar(filtered_data["Residency"], filtered_data["VALUE"], color='skyblue')
plt.title("Mean Nightly Accommodation Costs for April 2023")
plt.xlabel("Residency")
plt.ylabel("Cost (in Euro)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

From the chart, we can observe how the costs vary across different residencies, with visitors from "USA & Canada" having the highest mean nightly accommodation costs, followed closely by "Other Residencies".
